```python
# 랜덤 시드 설정
tf.random.set_seed(1)
```

```python
# 데이터를 읽어온다.
df1 = pd.read_csv('data/iris.csv', header=None)
df1.columns = ['꽃받침길이', '꽃받침너비', '꽃잎길이', '꽃잎너비', '품종']
df1.head()
```

```python
# 입력과 결과로 나눈다.
X = df1.drop('품종', axis=1)
y = df1['품종']

display(X)
display(y)
```

```python
# 문자열로 되어 있는 데이터를 숫자로 변환한다.
encoder1 = LabelEncoder()
encoder1.fit(y)
y = encoder1.transform(y)
y
```

```python
# 원핫인코딩.
# 분류의 경우에는 결과 데이터가 0부터 1씩 증가되는 형태라고 하더라도
# 연속성을 가진 데이터가 아닌 각각의 별개의 데이터를 의미한다.
# 딥러닝은 무조건 회귀를 한다. 이에 분류인 경우에는 결과 값을 보고 if 문으로
# 분기해서 결과를 우리가 직접 파악해야 한다. 하지만 원핫 인코딩을 쓰게 되면
# 결과데이터의 수 만큼 컬럼이 생성되고 각 컬럼은 0 아니면 1만 들어가 있게 
# 된다. 향후 최종 각 컬럼에 담겨져 있는 값중 가장 큰값의 인덱스를 가져오면
# 그것이 결과 데이터가 될 수 있다.
y_encoded = to_categorical(y)
y_encoded
```

#### 이진 분류
- 결과가 두 가지인 경우
- 출력층의 노드를 1로 설정한다.
- 출력층의 활성 함수는 sigmoid로 설정한다.
- compile시 손실함수는 binary_crossentropy로 설정한다.

#### 다중 분류
- 결과가 두 가지  이상인 경우(3가지 이상인 경우에 해주세요)
- 원핫 인코딩을 해야 하기 때문에 출력층의 노드 개수는 결과 데이터의 종류 수로 설정한다.
- 출력층의 활성 함수는 softmax로 설정한다.
- compile시 손실함수는 categorical_crossentropy로 설정한다.

```python
# 모델 설계
model = Sequential()

model.add(Dense(16, input_dim=4))
model.add(Activation('relu'))

# 다중 분류의 출력층의 노드 개수는 결과 데이터의 종류의 개수로 설정해야 한다.
model.add(Dense(3))
model.add(Activation('softmax'))
```

```python
# 모델 컴파일
# 다중 분류인 경우에는 손실함수를 categorical_crossentropy로 설정한다.
model.compile(loss='categorical_crossentropy', optimizer='adam',
              metrics=['accuracy'])
```

```python
# 학습한다.
model.fit(X, y_encoded, epochs=50, batch_size=10)
```

```python
# 예측 정확도
# 주어진 입력 데이터를 통해 예측 결과를 도출하고 설정한 결과 데이터와 
# 비교해서 얼마나 일치하는지 확인한다.
a1 = model.evaluate(X, y_encoded)
print(f'손실률 : {a1[0]}')
print(f'정확도 : {a1[1]}')
```
